### Abstract

The Movie Database (TMDb) is a community built movie and TV database. Every piece of data has been added by our amazing community dating back to 2008. Here, our aim is to extract movie data and reformat it to fit in a Conceptual database schema

### Data Description

The dataset is from Kaggle website. It contains two files. one is tmdb_5000_movies which contains 20 columns and 4803 records and other one is tmdb_5000_credits which contains 4 columns and 4803 records.
Below is the link tot the dataset from Kaggle

#### https://www.kaggle.com/tmdb/tmdb-movie-metadata

````python
#There are two datasets 
#Below are the columns in the tmdb_5000_movies dataset:
#•	budget #Total cost of the movie in which it is made.
#•	genre #Gives information about movie belongs in which all genre
#•	Id #Movie Id which is unique
#•	keywords #Keywords which define movie
#•	original_language #Original language of the movie
#•	overview #Overview of the movie
#•	popularity #Popularity of the movie
#•	production_companies #Company which has produced this movie
#•	production_countries #Countries in which movie is produced
#•	release_date #Date on which movie is released
#•	revenue #Revenue generated by the movie
#•	runtime #Movie’s runtime
#•	tagline #Movie’s tagline
#•	title #Title of the movie
#•	vote_average #Average number of votes movie received
#•	vote_count #Total number of votes movie received

#Below are the columns in the tmdb_5000_credits dataset:
#•	movie_id #Id of the movie
#•	title #Movie title
#•	Cast #Cast of the movie
#•	crew #Crew of the movie
````

#### Below are the steps to convert json object to dataframe

#### First converting genre json column into dataframe

#### step 1 : Create a empty list. Below I have created empty list  to collect genre data into "genre_column_list"

````python
genre_column_list =[]
````

#### step 2 : Appending genre json data into python list "genre_column_list". Using json.loads() method to parse data

````python
for each in movies_df['genres']:
    genre_column_list.append(json.loads(each))
````

#### Checking result in genre_column_list by displaying 5 records

````python
genre_column_list[:5]
````

#### step 3 : Create empty list to collect movieid

````python
movieid_genre_list = []
````

#### step 4 : Here, we are using built-in function enumerate which adds a counter to genere_column_list and returns enumerat object.
#### This enumerate object is used directly in for loop which returns a list of movieid's

````python
for i,x in enumerate(genre_column_list):
    for y in x:
        movieid_genre_list.append(movies_df['id'][i])
````

#### Below I am displaying 5 records in movieid_genre_list 

````python
movieid_genre_list[:5]
````

#### step 5 : Creating genre_list and genre_id_list to collect genre id and name which are the records of genre column

````python
genre_list = []
genre_id_list=[]
````

#### step 6 : Using for loop to append name and id in genre_list and genre_id_list respectively

````python
for x in genre_column_list:
    for y in x:
        genre_id_list.append(y['id'])
        genre_list.append(y['name'])           
````

#### Displaying 5 records in genre_list 

````python
genre_list[:5]
````

````python
genre_id_list[:5]
````

#### step 6 : Creating dataframe using pandas library which contains movie_id, genre_id and genre

````python
genre_df = pd.DataFrame(genre_list, columns=['Genre'])
genre_df['ID'] = pd.Series(genre_id_list)
genre_df['Movie_ID'] = pd.Series(movieid_genre_list)
genre_df = genre_df[['Movie_ID','ID', 'Genre']]
````

#### Below is the output displaying few records

````python
genre_df.head()
````

#### Similarly, lets extract "cast" json object into dataframe

#### Step 1 - Create Empty list to store cast column

#### step 3 : Create empty list to collect movieid¶

````python
movieid_cast_list = []
````

#### step 5 : Creating list for all the elements in cast column

````python
castid_list = []
character_list = []
creditid_list = []
gender_list = []
id_list = []
name_list = []
order_list = []
````
